```python
import pandas as pd

file_path = 'updatedDay.csv'
data = pd.read_csv(file_path)
```

```python
data
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
X
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train.value_counts()
```

```python
normal_indices = y_train[y_train == 0].index
malicious_indices = y_train[y_train != 0].index

normal_indices_sampled_10 = normal_indices[:int(len(normal_indices) * 0.1)]
normal_indices_sampled_20 = normal_indices[:int(len(normal_indices) * 0.2)]
normal_indices_sampled_50 = normal_indices[:int(len(normal_indices) * 0.5)]

sampled_indices_10 = normal_indices_sampled_10.append(malicious_indices)
sampled_indices_20 = normal_indices_sampled_20.append(malicious_indices)
sampled_indices_50 = normal_indices_sampled_50.append(malicious_indices)

X_train_sampled_10 = X_train.loc[sampled_indices_10]
y_train_sampled_10 = y_train.loc[sampled_indices_10]

X_train_sampled_20 = X_train.loc[sampled_indices_20]
y_train_sampled_20 = y_train.loc[sampled_indices_20]

X_train_sampled_50 = X_train.loc[sampled_indices_50]
y_train_sampled_50 = y_train.loc[sampled_indices_50]
```

```python
y_train_sampled_10.value_counts()
```

```python
normal_test_count = sum(y_test == 0)
malicious_test_count = sum(y_test != 0)

normal_indices_test = y_test[y_test == 0].index[:malicious_test_count]
malicious_indices_test = y_test[y_test != 0].index

balanced_indices_test = normal_indices_test.append(malicious_indices_test)

X_test_balanced = X_test.loc[balanced_indices_test]
y_test_balanced = y_test.loc[balanced_indices_test]
```

```python
y_test_balanced.value_counts()
```

```python
y_test.value_counts()
```

```python
import os

train_sampled_10 = pd.concat([X_train_sampled_10, y_train_sampled_10], axis=1)
train_sampled_20 = pd.concat([X_train_sampled_20, y_train_sampled_20], axis=1)
train_sampled_50 = pd.concat([X_train_sampled_50, y_train_sampled_50], axis=1)
test_balanced = pd.concat([X_test_balanced, y_test_balanced], axis=1)

folder_name = "sampled_data_dr/day"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

train_sampled_10.to_csv(os.path.join(folder_name, 'train_10.csv'), index=False)
train_sampled_20.to_csv(os.path.join(folder_name, 'train_20.csv'), index=False)
train_sampled_50.to_csv(os.path.join(folder_name, 'train_50.csv'), index=False)
test_balanced.to_csv(os.path.join(folder_name, 'test.csv'), index=False)
```
